```python
day = int(input("Enter a day (1-31): "))
month = int(input("Enter a month (1-12): "))

if month not in range(1, 13):
    print("Error. Month must be between 1 and 12.")
elif day not in range(1, 32):
    print("Error. Day must be between 1 and 31.")
elif month not in [1, 3, 5, 7, 8, 10, 12] and day == 31:
    print("Error. Day must be within the month.")
elif month == 2 and day > 29:
    print("Error. Day must be within the month.")
else:
    print("The date is valid.")
```

```python
# TODO: Add support for Saturn (9), Uranus (8.7), and Neptune (11).
planet = input("Pick a planet: Mars, Mercury, Venus, or Jupiter? ")

# We measure a planet's gravitional force in m/s^2.
gravity = 0.0
if planet == "Mars" or planet == "Mercury":
    gravity = 3.7
elif planet == "Venus":
    gravity = 8.9
elif planet == "Jupiter":
    gravity = 23.1
else:
    print("Unrecognized planet.")

if gravity:
    earth_weight = float(input("How much does it weigh on Earth (kg)? "))

    # An object's weight = mass * gravity.
    earth_gravity = 9.81
    mass = earth_weight / earth_gravity
    new_weight = round(mass * gravity, 1)

    print("It would weigh " + str(new_weight) + " kg on " + planet + ".")
```

Erick created a flashcard program to help him review Python concepts.
As he’s studying, he realizes that when he gets a question wrong, it would help if the program showed the correct answer.
That way, he can learn from his mistakes!

    For each question, if the user enters an incorrect answer, print the correct answer.

For example, if the user answers 4 for the second question, the program might print:
The correct answer is: 2.
Careful - if the user enters the correct answer, the program should still print:
You got it!, but it should not print: The correct answer is: 2.


```python
correct = 0
answer = input("What data type is the value 6.3? \n").strip().lower()
if answer == "float":
    print("You got it!")
    correct += 1
elif answer == 'integer' or 'int':
    print('Close! However Integers cannot have decimals.')
else:
    print("Sorry, the correct answer is float.")

answer = int(input("What does len(\"hi\") evaluate to? \n"))
if answer == 2:
    print("You got it!")
    correct += 1
else:
    print("Sorry, the correct answer is 2.")

print(f"You got a total of {correct} questions right.")
```
