# Average of a list of integers

Find the avgerage of test scores from user input, as user if they want to enter more before ending the program.


```python
# Enter scores
num_scores = int(input(f"Enter the Number of Scores you have: "))
scores = []
for i in range(num_scores):
    score = float(input(f"Enter Score {i + 1}: "))
    scores.append(score)

    # Calculate the average
    average = sum(scores) / len(scores)
# Print the average
print(f"The average of the scores is: {average:.2f}")

# Ask the user if they want to enter more scores
while True:
    more_scores = input(
        "Do you want to enter another score? (yes/no): ").strip().lower()
    if more_scores == 'yes':
        score = float(input(f"Enter Score {i + 1}: "))
        scores.append(score)
        # Calculate the average
        average = sum(scores) / len(scores)
        # Print the average
        print(f"The average of the scores is: {average:.2f}")
    elif more_scores == 'no':
        print(f'Good luck on your next test!')
        break
    else:
        print("Invalid input. Please enter 'yes' or 'no'.")
```
